```python
import pandas as pd
import xarray as xr
```

```python
nc_data = xr.open_dataset(r'D:\myStudy\Python\imd data converter\_Clim_Pred_LRF_New_RF25_IMD0p252019.nc')
nc_data
```

```python
nc_arr = nc_data.to_array()
nc_arr
```

```python
# Merge/stack lat lon columns to reduce 3d data to 2d data
nc_stacked = nc_arr.stack(z=("lon", "lat"))
nc_stacked
```

```python
# remove the 3rd variable and transpose the data
nc_stacked = nc_stacked.squeeze("variable")
nc_stacked = nc_stacked.transpose()
nc_stacked.shape
```

```python
# save as csv file
nc_stacked.to_pandas().to_csv(r'D:\myStudy\Python\imd data converter\201911.csv',index=True, encoding='utf-8')
```
